```python
import math 
import numpy as np 
import matplotlib.pyplot as plt

s = 750
h = 9

def f(x):
    y = (math.pi*x)*(h**2 + x**2)**0.5 - s
    return y

print("x \t f(x)")
tabla = []
for n in range(21):
    y = f(n)
    print(n, y)

x = np.linspace(0, 20, 20)
plt.plot(x, f(x))
plt.grid()
plt.show()
```

```python
a = 10
b = 20
r = (a+b)/2
error = 1
tol = 1e-6
i = 0
Error = []

print("Método de la Bisección")
print()
print("{0:10s}{1:16.9s}{2:16.9s}{3:16.9s}{4:16.9s}{5:16.9s}{6:16.9s}{7:16.9s}".format("i","a","b","f(a)","f(b)","r","f(r)","Error"))

while (error > tol):
    print("{0:3d}{1:16.9f}{2:16.9f}{3:16.9f}{4:16.9f}{5:16.9f}{6:16.9f}{7:16.9f}".format(i,a,b,f(a),f(b),r,f(r),error))  
    if (f(r)*f(a) < 0):
        b = r
    elif f(r)*f(b) < 0:
        a = r
    else:
        break
    rold = r
    r = (a+b)/2
    error = abs((r-rold)/r)
    i = i + 1
    Error.append(error)
print()
print("La raiz encontrada por el método de Bisección es: ", r, "con un error de:", error)
print("La cantidad de iteraciones necesarias es de: ", i-1)

plt.plot(Error)
```

```python
a = 10
b = 20
r = b - (f(b)*(a-b))/(f(a)-f(b))
error = 1
tol = 1e-6
i = 0
Error = []

print("Método de la Regla Falsa")
print()
print("{0:10s}{1:16.9s}{2:16.9s}{3:16.9s}{4:16.9s}{5:16.9s}{6:16.9s}{7:16.9s}".format("i","a","b","f(a)","f(b)","r","f(r)","Error"))

while (error > tol):
    print("{0:3d}{1:16.9f}{2:16.9f}{3:16.9f}{4:16.9f}{5:16.9f}{6:16.9f}{7:16.9f}".format(i,a,b,f(a),f(b),r,f(r),error))  
    if (f(r)*f(a) < 0):
        b = r
    elif f(r)*f(b) < 0:
        a = r
    else:
        break
    rold = r
    r = b - (f(b)*(a-b))/(f(a)-f(b))
    error = abs((r-rold)/r)
    i = i + 1
    Error.append(error)
print()
print("La raiz encontrada por el método de Regla Falsa es: ", r, "con un error de:", error)
print("La cantidad de iteraciones necesarias es de: ", i)

plt.plot(Error)



```

```python
def fprima(x):
    y_ = (math.pi)*(h**2 + 2*x**2)*(h**2 + x**2)**(-0.5)
    return y_

a = 10
error = 1
tol = 1e-6
i = 0
r = a - f(a)/fprima(a)
Error = []

print("Método de Newton Raphson")
print()
print("{0:10s}{1:16.9s}{2:16.9s}{3:16.9s}{4:16.9s}{5:16.9s}".format("i","a","f(a)","r","f(r)","Error"))

while(error > tol):
    print("{0:3d}{1:16.9f}{2:16.9f}{3:16.9f}{4:16.9f}{5:16.9f}".format(i,a,f(a),r,f(r),error))
    a = r
    r = a - f(a)/fprima(a)
    error = abs((r-a)/r)
    i = i + 1
    Error.append(error)

print()
print("La raiz encontrada por el método de Newton Raphson es: ", r, "con un error de:", error)
print("La cantidad de iteraciones necesarias es de: ", i)

plt.plot(Error)
    
```
